```python
# histogram function
def histo(visual_rows, bins):

    fraud_counts = visual_rows.value_counts()
    fraud_percentages = (fraud_counts / len(visual_rows)) * 100
    # Plot the histogram
    plt.figure(figsize=(6,3))
    plt.hist(visual_rows, bins=bins, edgecolor='black')
    # Add labels and title
    plt.xlabel('isFraud (True/False)')
    plt.ylabel('Frequency')
    plt.title('Histogram of Fraud')

    for i, count in enumerate(fraud_counts):
        percentage = fraud_percentages[i]
        plt.text(i, count + 5, f'{percentage:.2f}%', ha='left', fontsize=12)


    # Show the plot
    plt.show()
```

2. Transaction amounts importance analysis.

```python
data_sorted = data.sort_values(by=['creditLimit', 'customerId', 'transactionDateTime'])
```

```python
data_sorted.head()
```

```python
# Feature: customerId
fea = "creditLimit"

# Group data by customerId and isFraud, and count occurrences
ans = (data_sorted[[fea, "isFraud"]].groupby([fea, "isFraud"], dropna=False).size()
       .reset_index(name="num_events")
       .pivot(index=fea, columns="isFraud", values="num_events"))

# Rename the columns
ans.columns = ["num_nonfraud", "num_fraud"]

# Fill NaN values (if a customer has only fraudulent or only legitimate transactions)
ans = ans.fillna(0)

# Plot a stacked bar chart
ax = ans.plot.bar(stacked=True, figsize=(10, 4))

# Calculate the fraud rate
ans["fraud_rate"] = np.round(ans["num_nonfraud"] / (ans["num_nonfraud"] + ans["num_fraud"]), 3)

# Add fraud rate labels on the bars
ax.bar_label(ax.containers[0], labels=ans["fraud_rate"].values)

# Set plot labels and title
ax.set_ylabel("# of transactions")
ax.set_title("Fraud Detection: Fraud Rate by Credit Limit")

# Show the plot
plt.show()
```

```python
# Filter rows where 'isFraud' is True
fraud_transactions = data_sorted[data_sorted["isFraud"] == True]
non_fraud_transactions = data_sorted[data_sorted["isFraud"] == False]

# Extract the 'transactionAmount' column from the filtered rows
fraud_transaction_amounts = fraud_transactions["transactionAmount"]
non_fraud_transaction_amounts = non_fraud_transactions["transactionAmount"]

plt.figure(figsize=(12, 4))  # Adjust the figure size for a wider layout
# Plot for 'isFraud == True'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(fraud_transaction_amounts, bins=100, edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Histogram of fraud')

# Plot for 'isFraud == False'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.hist(non_fraud_transaction_amounts, bins=100, edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Histogram of nonFraud')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

Fraud and non frand transactions have different distributions.

```python
# Statistical analysis
from scipy.stats import ks_2samp, ttest_ind

fraud_trans_mean, nonfraud_trans_mean = np.mean(fraud_transaction_amounts), np.mean(non_fraud_transaction_amounts)
fraud_trans_std, nonfraud_trans_std = np.std(fraud_transaction_amounts), np.std(non_fraud_transaction_amounts)
fraud_trans_median, nonfraud_trans_median = np.median(fraud_transaction_amounts), np.median(non_fraud_transaction_amounts)
fraud_trans_range, nonfraud_trans_range = (np.min(fraud_transaction_amounts), np.max(fraud_transaction_amounts)), (np.min(non_fraud_transaction_amounts), np.max(non_fraud_transaction_amounts))

# Statistical tests
# Kolmogorov-Smirnov Test (checks if two distributions differ significantly)
ks_stat, ks_pval = ks_2samp(fraud_transaction_amounts, non_fraud_transaction_amounts)

# T-test (checks if means are significantly different)
tt_stat, tt_pval = ttest_ind(fraud_transaction_amounts, non_fraud_transaction_amounts)

print("Statistical Properties:")
print(f"Distribution 1 - Mean: {fraud_trans_mean:.2f}, Std Dev: {fraud_trans_std:.2f}, Median: {fraud_trans_median:.2f}, Range: {fraud_trans_range}")
print(f"Distribution 2 - Mean: {nonfraud_trans_mean:.2f}, Std Dev: {nonfraud_trans_std:.2f}, Median: {nonfraud_trans_median:.2f}, Range: {nonfraud_trans_range}")
print("\nStatistical Tests:")
print(f"KS Test: Statistic = {ks_stat:.3f}, P-value = {ks_pval:.3f}")
print(f"T-Test: Statistic = {tt_stat:.3f}, P-value = {tt_pval:.3f}")
if ks_pval < 0.05:
    print("\nThe two distributions are significantly different based on the KS test (p < 0.05).")
else:
    print("\nThe two distributions are not significantly different based on the KS test (p â‰¥ 0.05).")

if tt_pval < 0.05:
    print("The means of the two distributions are significantly different based on the T-test (p < 0.05).")
else:
    print("The means of the two distributions are not significantly different based on the T-test (p â‰¥ 0.05).")
```

Both right-sknew. Fraud transaction distribution is less spread.
